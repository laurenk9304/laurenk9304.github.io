### Top 5 Hours for Ask Posts Comments

```python
print('Top 5 Hours for Ask Posts Comments')

for avg in sorted_swap[:5]:
    #USA (EST UTC -5) Time Zone 
    swap_strp = dt.datetime.strptime(avg[1], '%H')
    swap_stfp = swap_strp.strftime('%H:%M')
    
    #Korea (UTC +9) Time Zone
    #Convert USA to Korea time zone 
    our_swap_strp = dt.datetime.strptime(avg[1], '%H') + dt.timedelta(hours = 14)
    our_swap_stfp = our_swap_strp.strftime('%H:%M')
    
    print('{est_time} EST or {our_time} Korea Time: {avg:.2f} average comments per post'. 
     format(est_time = swap_stfp, our_time = our_swap_stfp, avg = avg[0]))
```

You should post between 3:00 pm - 4:00 pm EST or 5:00 am Korea time to receive the most average comments per post. This means that Hacker News is probably based in the US. People on average check their phones during the mid afternoon for any updates.
